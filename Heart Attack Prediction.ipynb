```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression 
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.neighbors import  KNeighborsClassifier
from sklearn.metrics import mean_squared_error
```

```python
y=data['output']
X=data.drop('output',axis=1)
```

```python
y.shape, X.shape
```

```python
y.head()
```

```python
X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
lr = LogisticRegression()
lr.fit(X_train, y_train)
accuracy_lr = lr.score(X_test, y_test)
accuracy_lr
```

```python
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
accuracy_knn = knn.score(X_test, y_test)
accuracy_knn
```

```python
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
accuracy_rf = rf.score(X_test, y_test)
accuracy_rf
```

```python
ad = AdaBoostClassifier()
ad.fit(X_train, y_train)
accuracy_ad = ad.score(X_test, y_test)
accuracy_ad
```

```python
prediction_lr = lr.predict(X_test)
prediction_lr
```

```python
mean_sqr_error = mean_squared_error(y_test, prediction_lr)
mean_sqr_error
```

```python
model=['Logistic Regression', 'Random Forest', 'KNeighbours', 'AdaBoost']
pd.DataFrame({'Model':model,'Accuracy':[round(accuracy_lr,2),round(accuracy_rf,2),round(accuracy_knn,2),round(accuracy_ad,2)],
              'Mean Square Error':[mean_squared_error(y_test, lr.predict(X_test)),
                                   mean_squared_error(y_test, rf.predict(X_test)),
                                   mean_squared_error(y_test, knn.predict(X_test)),
                                   mean_squared_error(y_test, ad.predict(X_test))]
              }).set_index('Model')
```
